```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
hpi_path = r"..\data\hpi.csv"
```

```python
hpi_data = pd.read_csv(hpi_path)

hpi_data.head()
```

```python
df = hpi_data.copy()
```

```python
df["Y"] = df["Year"].apply(lambda y: int(y.split("-")[0]))
```

```python
df.describe()
```

```python
cities = ['Bangalore', 'Chennai', 'Delhi',
       'Jaipur', 'Kanpur', 'Kochi', 'Kolkata', 'Lucknow', 'Mumbai']

for city in cities:
    minimum = df[city].min()
    maximum = df[city].max()
    y_min = df.loc[df[city]==minimum, "Y"].values[0]
    y_min_q = df.loc[(df[city]==minimum) & (df["Y"]==y_min), "Quarter_Num"].values[0]
    y_max = df.loc[df[city]==maximum, "Y"].values[0]
    y_max_q = df.loc[(df[city]==maximum) & (df["Y"]==y_max), "Quarter_Num"].values[0]
    print(f"City: {city}\nMin HPI: {minimum} in year {y_min} Quarter {y_min_q}, Max HPI: {maximum} in year {y_max} Quarter {y_max_q}.", end=" ")
    if y_min < y_max:
        print(f"Jump = {maximum/minimum:.2f}x")

    print("-"*92) 
```

```python
df.columns
```

```python
df2 = df[[
    'Y', 'Quarter_Num', 'Bangalore', 'Chennai', 'Delhi',
       'Jaipur', 'Kanpur', 'Kochi', 'Kolkata', 'Lucknow', 'Mumbai',
       'All India'
]]
```
